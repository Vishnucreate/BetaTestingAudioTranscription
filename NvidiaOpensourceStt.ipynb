```python
# Cell 5: Build NeMo manifest JSONs from the .srt files
import srt, json

def build_manifest(prefix):
    in_srt = f"{prefix}.en.srt"
    out_json = f"{prefix}_manifest.json"
    with open(in_srt, "r", encoding="utf-8") as f, open(out_json, "w", encoding="utf-8") as o:
        subs = list(srt.parse(f.read()))
        for sub in subs:
            dur = round((sub.end - sub.start).total_seconds(), 2)
            if dur > 0.3:
                o.write(json.dumps({
                    "audio_filepath": f"{prefix}.wav",
                    "offset": round(sub.start.total_seconds(), 2),
                    "duration": dur,
                    "text": sub.content.replace("\n", " ")
                }) + "\n")
    print("✔︎", out_json)

build_manifest("video1")
build_manifest("video2")
```

```python
# Cell 6: Combine into train/val manifests
!cat video1_manifest.json > train_manifest.json
!cat video2_manifest.json > val_manifest.json
```

```python
# Cell 7: Fine-tune QuartzNet15x5Base on your football data
import nemo.collections.asr as nemo_asr
from omegaconf import OmegaConf

model = nemo_asr.models.EncDecCTCModel.from_pretrained("QuartzNet15x5Base-En")

cfg = OmegaConf.create({
  "model": {
    "train_ds": {
      "manifest_filepath": "train_manifest.json",
      "sample_rate": 16000,
      "labels": list(" abcdefghijklmnopqrstuvwxyz'"),
      "batch_size": 8,
      "shuffle": True,
      "num_workers": 2
    },
    "validation_ds": {
      "manifest_filepath": "val_manifest.json",
      "sample_rate": 16000,
      "labels": list(" abcdefghijklmnopqrstuvwxyz'"),
      "batch_size": 8,
      "shuffle": False,
      "num_workers": 2
    },
    "optim": {"lr": 0.001}
  },
  "trainer": {"max_epochs": 10, "gpus": 1}
})

model.setup_training_data(train_data_config=cfg.model.train_ds)
model.setup_validation_data(val_data_config=cfg.model.validation_ds)
model.train()
model.save_to("quartznet_football.nemo")
```

```python
# Cell 8: Inference on video2.wav (unseen during training)
transcripts = model.transcribe(["video2.wav"])
print("▶ Transcript of video2.wav:\n", transcripts[0])
```
